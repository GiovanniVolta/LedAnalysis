```python
folder = '/dali/lgrandi/giovo/XENONnT/commissioning/Data/20200422/'
np_load_old = np.load
np.load = lambda *a,**k: np_load_old(*a, allow_pickle=True, **k)
data1 = np.load(folder + 'Amplitude_007333.npz')
data2 = np.load(folder + 'Amplitude_007335.npz')
np.load = np_load_old
```

```python
amplitude3_6 = data1['x']
amplitude4_0 = data2['x']
```

```python
channels = np.arange(0, 494, 1)
```

#### LED voltage @ 3,6 V

```python
bad_ch = [ ]
for ch in channels:
    try:
        _, _, check = PMT.get_speinput(amplitude3_6, ch)
        print('PMT: ', ch, ' - ', check)
    except:
        print('Bad: ', ch)
        bad_ch.append(ch)
```

```python
good_ch = [ ]
for ch in channels:
    if ch not in bad_ch:
        good_ch.append(ch)
print(bad_ch, good_ch)
```

```python
df_fit_3_6, df_led_3_6, window3_6 = PMT.get_ledwindow(amplitude3_6 , good_ch)
```

```python
PMT.plot_ledwindow(df_led_3_6, lastchannel=494)
```
